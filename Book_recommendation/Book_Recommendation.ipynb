```python
import numpy as np
import pandas as pd
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
```

```python
books = pd.read_csv('Books.csv')
users = pd.read_csv('Users.csv')
ratings = pd.read_csv('Ratings.csv')
```

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
user_rating_counts = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
frequent_users = user_rating_counts[user_rating_counts].index
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(frequent_users)]
```

```python
book_rating_counts = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
popular_books = book_rating_counts[book_rating_counts].index
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(popular_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating').fillna(0)
```

```python
pt
```

```python
svd = TruncatedSVD(n_components=15, random_state=42)
matrix_reduced = svd.fit_transform(pt)
```

```python
similarity_scores = cosine_similarity(matrix_reduced)
```

```python
def recommend(book_name, number_of_books=5):
    if book_name not in pt.index:
        return "Book not found in the dataset."
    index = np.where(pt.index == book_name)[0][0]

    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:number_of_books]

    recommendations = []
    for item in similar_items:
        book_info = books[books['Book-Title'] == pt.index[item[0]]].drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M']]
        recommendations.append(book_info.values[0])

    return pd.DataFrame(recommendations, columns=['Book-Title', 'Book-Author', 'Image-URL-M'])
```

```python
from IPython.display import Image, display

def display_recommended_images(recommendations_df):
    for index, row in recommendations_df.iterrows():
        title = row['Book-Title']
        image_url = row['Image-URL-M']
        print(f"{title}")
        display(Image(url=image_url))
        print("\n")
```

```python
recommendations = recommend('Year of Wonders', 7)

if isinstance(recommendations, pd.DataFrame):
    display_recommended_images(recommendations)
else:
    print(recommendations)
```
